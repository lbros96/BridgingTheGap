```python
import torch

import torchaudio

import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
import librosa

import pandas as pd

import utils
```

```python
audio = torchaudio.load("wavenet_output/Patch3-Example01.wav.wav")

print(audio)
audio = audio[0]
audio = audio.numpy()

librosa_melspec = librosa.feature.melspectrogram(y=audio[0], sr=16000, n_fft=1024, n_mels = 80, hop_length=128)

# Let's look at the spectrogram of one example
plt.imshow(librosa.power_to_db(librosa_melspec, ref=np.max), aspect='auto', origin='lower')
plt.colorbar()
```

```python
ipd.Audio(audio[0], rate=16000)
```
